# Data Augmentation

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(rotation_range=10,zoom_range = 0.1,
        width_shift_range=0.1,height_shift_range=0.1,
        horizontal_flip=False,vertical_flip=False,
        preprocessing_function=HE)

```

```
batch_size=128
Training_data = datagen.flow(trainX,
                             y=trainY,
                            batch_size = batch_size)
```

```
Validation_data = datagen.flow(validX,
                             y=validY,
                            batch_size = batch_size)
Test_data = datagen.flow(testX,
                             y=testY,
                            batch_size = batch_size)
```

# Callbacks

```
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint,ReduceLROnPlateau
es = EarlyStopping(monitor='val_loss', mode='auto', verbose=1, patience=1)
mc = ModelCheckpoint(filepath="best_weights.hdf5", monitor = 'val_accuracy',verbose=1,save_best_only=True)
lrr = ReduceLROnPlateau(monitor='val_accuracy',factor=.01,patience=2,min_lr=1e-5)
```

# Convolution neural network building (finding the best model)

```
from tensorflow.keras.layers import Conv2D, MaxPooling2D,AveragePooling2D, Flatten, Dense, Dropout, Activation , Concatenate, Input , BatchNormalization
```

```
def build_model():
  model=keras.models.Sequential()
  model.add(Conv2D(32,(5,5), activation='relu',padding='same',name = 'Conv1_3',input_shape=trainX.shape[1:]))
  model.add(BatchNormalization(name='norm1'))
  model.add(MaxPooling2D((2,2),name = 'MaxPool1'))
  model.add(Conv2D(64,(3,3), activation='relu', padding='same',name= 'Conv3_3'))
  model.add(BatchNormalization(name='norm9'))
  model.add(AveragePooling2D((2,2),name = 'AvgPool1'))
  model.add(Conv2D(128,(1,1), activation='relu', padding='same',name= 'Conv5_1'))
  model.add(BatchNormalization(name='norm4'))

  model.add(Flatten(name = 'Flatten'))


  model.add(Dense(units = 120, activation = 'relu'))
  model.add(Dropout(rate = 0.2))
  model.add(BatchNormalization(name='norm6'))
  model.add(Dense(units = 90, activation = 'relu'))
  model.add(BatchNormalization(name='norm7'))
  model.add(Dense(units = 65, activation = 'relu'))
  model.add(BatchNormalization(name='norm8'))
  model.add(Dense(units = 43, activation = 'softmax' , name='preds'))
  return model
```

```
model=build_model()
```

```
model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.01,momentum=0.9), loss='categorical_crossentropy', metrics=['accuracy'])
```

```
history = model.fit(Training_data,epochs= 5,validation_data= Validation_data,verbose=1,callbacks=[es , mc,lrr],class_weight=classWeight )
```
